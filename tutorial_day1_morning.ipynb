```python
# Load the CSV (no header)
pythia = pd.read_csv('data/herwig.csv', header=None)

# Extract columns at index 1, 2, 3, 4 (2nd through 5th entries)
Mult = pythia.iloc[:, 0:1].values  
JetMass = pythia.iloc[:, 1:2].values
JetpT = pythia.iloc[:, 2:3].values
JetX = pythia.iloc[:, 3:4].values
Jetzg = pythia.iloc[:, 4:5].values
JetRg = pythia.iloc[:, 5:6].values
JetkT = pythia.iloc[:, 6:7].values
Jetmg = pythia.iloc[:, 7:8].values
```
